```python
import numpy as np
import pandas as pd
```

```python
median_gain_5dd_performances = pd.read_csv('data/stock/03_model/performance/medianGain/5dd.csv').dropna()
max_loss_5dd_performances = pd.read_csv('data/stock/03_model/performance/maxLoss/5dd.csv').dropna()
```

```python
median_gain_5dd_forecasts = pd.read_csv('data/stock/04_forecast/medianGain/5dd.csv').dropna()
max_loss_5dd_forecasts = pd.read_csv('data/stock/04_forecast/maxLoss/5dd.csv').dropna()
```

# 5 dd Analysis

```python
merged_win_data =  pd.merge(
                            median_gain_5dd_forecasts[median_gain_5dd_forecasts['Forecast High Gain 5dd'] >= 0.5].sort_values('Forecast High Gain 5dd', ascending=False),
                            median_gain_5dd_performances[median_gain_5dd_performances['Test - Gini'] >= 0.5],
                            on='Kode',
                            how='inner'
                        )

filtered_win_data = merged_win_data[merged_win_data['Threshold'] > 0] \
                        .reset_index(drop=True)

selected_win_emiten = filtered_win_data['Kode'].values
```

```python
filtered_win_data.head(20)
```

```python
merged_lose_data =  pd.merge(
                                max_loss_5dd_forecasts[max_loss_5dd_forecasts['Kode'].isin(selected_win_emiten)],
                                max_loss_5dd_performances,
                                on='Kode',
                                how='inner'
                            ) \
                        .sort_values('Threshold', ascending=False) \
                        .reset_index(drop=True)
```
