```python
import numpy as np
import pandas as pd
```

```python
def dd_analysis(days):
    median_gain_dd_performances = pd.read_csv(f'data/stock/03_model/performance/medianGain/{days}dd.csv').dropna()    
    median_gain_dd_forecasts = pd.read_csv(f'data/stock/04_forecast/medianGain/{days}dd.csv').dropna()

    merged_win_data =  pd.merge(
                                median_gain_dd_forecasts[median_gain_dd_forecasts[f'Forecast High Gain {days}dd'] >= 0.5].sort_values(f'Forecast High Gain {days}dd', ascending=False),
                                median_gain_dd_performances[median_gain_dd_performances['Test - Gini'] >= 0.3],
                                on='Kode',
                                how='inner'
                            )
    
    
    filtered_win_data = merged_win_data[merged_win_data['Threshold'] > 0] \
                            .reset_index(drop=True)
    
    selected_win_emiten = filtered_win_data['Kode'].values

    performance_summary = median_gain_dd_performances.describe()
    
    return filtered_win_data, performance_summary
```

# 5dd Analysis

```python
filtered_win_data, performance_summary = dd_analysis(5)
```
